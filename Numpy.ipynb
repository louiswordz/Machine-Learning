## Correlation
* R = 1 -> Strong positive relationship
* R = 0 -> Not linearly correlated
* R = -1 -> Strong negative relationship

## Pearson Correlation : Assumptions
* Your data is normally distributed
* You have continuous, numeric variables.
* Your variables are linearly related

## Use Pearson correlation
* To uncover (linear) relationships between variables

* Not to rule out possible (nonlinear) relationships between variables

## Spearman's Rank Correlation
This finds the R correlation between variable-pair of ordinal data types. Variable-pair are then able o be ranked according to the strength of the correlation between them.

* R = 1 -> Strong positive relationship
* R = 0 -> Not linearly correlated
* R = -1 -> Strong negative relationship


* Your variable are ordinal (in other words numerical, but able to be ranked like categorical variable).

* your variable are related  nonlinearly.

* Your data is non-normally distributed.


##### Chisquqre Tables test for independence
* p<0.05 -> Rejected null hypothesis and conclude that the variable are correlated

* p >0.05 -> Accept null hypothesis and conclude that the variable are independent

### When to Use Chi-square Tables
* Your variables are categorical or numerical

* You have binned the numerical variables

### Chi-square test for independence

```python
table = pd.crosstab(cyl, am)

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table.values)
print("Chi-square  Statistics %.3f p_value %.3f Degree of Freedom %.3f" %(chi2, p, dof))

```

```python
table = pd.crosstab(cyl, gear)

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table.values)
print("Chi-square  Statistics %.3f p_value %.3f" %(chi2, p))
```
